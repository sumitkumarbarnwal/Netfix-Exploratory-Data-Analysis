```python
df.head()
```

```python
def catigorize_col (df, col, labels):
 edges = [df[col].describe()['min'],
 df[col].describe()['25%'],
 df[col].describe()['50%'],
 df[col].describe()['75%'],
 df[col].describe()['max']]
 df[col] = pd.cut(df[col], edges, labels = labels, duplicates='drop')
 return df
```

```python
labels = ['not_popular', 'below_avg', 'average', 'popular']
catigorize_col(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```

```python
most_watched = df.groupby('Title')['Popularity'].sum().sort_values(ascending=False)

# Create a horizontal bar plot
plt.figure(figsize=(12, 8))
sns.barplot(x=most_watched.values[:20], y=most_watched.index[:20], palette="viridis")
plt.title('Most Watched Movies (Top 20)')
plt.xlabel('Total Popularity Score')
plt.ylabel('Movie Title')
plt.show()
```
